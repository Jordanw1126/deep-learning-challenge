## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
new_app_df = application_df.drop(columns=['EIN','NAME'])
new_app_df
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
unique_counts = new_app_df.nunique()
unique_counts
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
#  YOUR CODE GOES HERE
value_counts = new_app_df['APPLICATION_TYPE'].value_counts()
value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
cutoff_value = 200

application_types_to_replace = value_counts[value_counts < cutoff_value].index.tolist()
# Replace in dataframe
for app in application_types_to_replace:
    new_app_df['APPLICATION_TYPE'] = new_app_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
new_app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
#  YOUR CODE GOES HERE
new_app_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
classification_counts = new_app_df['CLASSIFICATION'].value_counts()
filtered_counts = classification_counts[classification_counts > 1]
filtered_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
cutoff_value = 1000
value_counts = new_app_df['CLASSIFICATION'].value_counts()
classifications_to_replace = value_counts[value_counts < cutoff_value].index.tolist()
# Replace in dataframe
for cls in classifications_to_replace:
    new_app_df['CLASSIFICATION'] = new_app_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
new_app_df['CLASSIFICATION'].value_counts()
```

```python
new_app_df.info()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
df_dummies = pd.get_dummies(new_app_df, columns=['APPLICATION_TYPE', "AFFILIATION", 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS'], drop_first=True)
df_dummies
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
X = df_dummies.drop(columns=['IS_SUCCESSFUL'], axis = 1)
y = df_dummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=78, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

hidden_nodes_layer1 = 80  # Number of nodes in the first hidden layer
hidden_nodes_layer2 = 30 # Number of nodes in the second hidden layer
nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation='relu', input_dim=36))
# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))
# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
